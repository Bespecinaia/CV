```python
import cv2
import numpy as np
```

```python
img = cv2.imread('Lenna.png')
```

```python
new_image = np.zeros(img.shape, img.dtype)
```

```python
contrast = 5
brighness = 2

for y in range(img.shape[0]):
    for x in range(img.shape[1]):
        for c in range(img.shape[2]):
            new_image[y,x,c] = np.clip(contrast * img[y,x,c] + brighness, 0, 255)
```

```python
io.imshow(new_image)
```

```python
test_img = np.clip(img * contrast + brighness, 0, 255)
```

```python
io.imshow(test_img)
```

```python
cv2.imwrite('Lenna_opencv.png', test_img)
```
